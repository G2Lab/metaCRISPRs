```python
import pandas as pd
```

```python
db = pd.read_csv('master_ref.csv')
db_1 = db['Sample']
a = db_1.tolist()
print(a)
```

```python
csv_filename = "locs.csv"
import requests
from bs4 import BeautifulSoup
import csv
with open(csv_filename, mode="w", newline="") as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(["Sample ID","Geographic Location", "Sample Description", "Longitude","Latitude"])  # Write header row
    for i in a:
        sample_id = i
        api_url = f"https://www.ebi.ac.uk/metagenomics/api/v1/samples/{sample_id}"
        response = requests.get(api_url)
        if response.status_code!=200:
            writer.writerow(sample_id)
            continue
        else:
            data = response.json()
            attributes = data['data']['attributes']['sample-metadata']
            geolocation_value = None

            for attribute in attributes:
                if attribute['key'] == 'geographic location (country and/or sea,region)':
                    geolocation_value = attribute['value']
                    break
        
            sample_desc = data['data']['attributes']['sample-desc']
            longitude = None
            for attribute in attributes:
                if attribute['key'] == 'geographic location (longitude)':
                    longitude = attribute['value']
                    break
        
            latitude = None
            for attribute in attributes:
                if attribute['key'] == 'geographic location (latitude)':
                    latitude = attribute['value']
                    break            
            print(longitude)
            print(latitude)
    
            writer.writerow([sample_id, geolocation_value, sample_desc, longitude, latitude])
```
